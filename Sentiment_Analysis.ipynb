# Sentiment Analysis of hotel reviews

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Arch0/Sentiment-Analysis-of-hotel-review/main/Data%20Analyst%20-%20Test%20Data.csv",encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
train = df[df['date'] < '20190710']
test = df[df['date'] > '20190820']
```

```python
# Removing punctuations
data=train.iloc[:,0]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

```

```python
# Convertng Reviews to lower case
data=data.str.lower()
data
```

```python
reviews=[]
reviews.append(" ".join(str(x) for x in data))
```

```python
reviews
```

```python
reviews[0]
```

```python
reviews=str(reviews)
```

```python
from wordcloud import WordCloud
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
wordcloud = WordCloud().generate(reviews)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
